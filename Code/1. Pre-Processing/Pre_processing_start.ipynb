```python
import gzip

def parse(path):
  g = gzip.open(path, 'rb')
  for l in g:
    yield eval(l)

def getDF(path):
  i = 0
  df = {}
  for d in parse(path):
    df[i] = d
    i += 1
  return pd.DataFrame.from_dict(df, orient='index')

df = getDF('\meta_Electronics.json.gz')
```

```python
print(df.shape)
print(df.columns)
print(df.info())
```

```python
product_reviews=pd.merge(df1,df,on='asin',how='left')
```
